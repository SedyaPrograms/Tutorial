+ **Метрика Accuresy**

  Дать выборку значений, ответы которой известны и посчитать процент угаданных
тогда метрику можно определить как:

  Accuresy = SUM( ( Y_pred(i) == Y_true(i) )/n), 

  i = [1,n], сумма по всем элиментам i

  ее значение от 0 (все не угадали) до 1(все угадали), тут все круто, если набор кошек и собак равновероятный (Accuresy = 0,5)

  Пусть есть пациент. Варианты: болен, здоров. У нас есть классификатор с 
Accuresy = 0,95. Почему это плохо? - Если модель, говорит, что все здоровы. Люди заходят в основном потестить модель и были здоровы (их 95%), в тоже время больные люди (5%) получают ответ, что они здоровы. это мб критично. Нужна другая метрика, которая ожидает несбалансированность классов, например учитывать веса.

**Разница между метрикой и функцией потерь**

Метрикой может являтся Recall или Precision или их комбинация, для выбора лучшей модели

Пусть мы хотим модель Машинного Обучения, которая эливантно предсказывает какие-нибудь товары на сайте, т.е. конечная цель, чтобы пользователи кликали чаще на опрделенные товары или любые примеры сложно формализуемые, например, как формализовать, что два изображения похожи. Поэтому часто пребегают к упрощению использую различные формулы, которые помогают максимизировать нашу метрику на основе функции потерь, т.е при оптимизации функции потерь, метрика должна тоже оптимизировать. Метрики не удобно пускать через град спуск (они не диффиринцируемы и не гладкие), а функцию потерь вполне себе можем.   

Таким образом:
*   Метрика используется на этапе, когда выбираем какая модель лучше
*   Функция потерь на обучении, чтобы оптимизировать модель
